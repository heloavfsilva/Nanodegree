```python
# import the necessary dependencies to perform the gathering, assessing, cleaning and plotting the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<a id='wrangling'></a>
## Data Wrangling

During the next cells we will load the data, check for duplicates and trim it, 
in order to be ready for analyze.


### General Properties

```python
# Load the data and print the data to analyze the information
df_movie = pd.read_csv('tmdb-movies.csv')
df_movie.head(5)
```

```python
# print the information of the dataset in order to look for missing values and duplicates
df_movie.info()
print('Count of duplicated: ' + str(df_movie.duplicated().sum()))
print('Columns and rows: ' + str(df_movie.shape))
```

### Remove Missing Data 

```python
#  clean the missing values and print again 
df_movie.dropna(inplace=True)
df_movie.info()
print('Count of duplicated: ' + str(df_movie.duplicated().sum()))
print('Columns and rows: ' + str(df_movie.shape))
```

```python
# Generate a new dataframe to split the column genres
new_df = pd.DataFrame(df_movie.genres.str.split('|').tolist(), index=df_movie.id).stack()
new_df = new_df.reset_index([0, 'id'])

new_df.columns = ['id', 'genres_split']
new_df.head()
```

```python
# Combine the genres dataframe with the Movie dataframe using an inner join relation
# Drop the previous column
df_combined = pd.merge(new_df, df_movie, on='id', how='inner')
df_combined.drop(columns="genres", inplace=True)
df_combined.head()
```

<a id='eda'></a>
## Exploratory Data Analysis


### 1. What is the correlation between the genres and the movies popularity mean? 

```python
# Print the Popularity data to explore it
df_combined['popularity'].describe()
```

```python
# Grouping the popularity by genres_split and calculate the mean 
# Plotting the data we can notice that the genres Adventure, Fantasy and Science Fiction has the highest Popularity Rating 
popularity_values=df_combined.groupby(['genres_split'])['popularity'].mean().plot(kind='bar')
plt.title('Correlation between Genres and Movies Popularity')
plt.xlabel('Genres Category')
plt.ylabel('Popularity Rating Mean');
```

### 2. What are the features of the most popular movies?

```python
# Checking for most popular movies data
df_top = df_combined.query('genres_split in ["Adventure","Fantasy","Science Fiction"]')
df_top.head()
```

```python
# Plotting the information of the top movie genres, was not possible to notice any positive correlation
pd.plotting.scatter_matrix(df_top, alpha=0.2, figsize=(10, 10), diagonal='kde');
```
